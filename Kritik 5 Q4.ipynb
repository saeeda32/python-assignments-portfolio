```python
import numpy as np
import matplotlib.pyplot as plt

def normal_density(mean, variance, x):
    f_sub_x = (1/np.sqrt(2*np.pi*variance))*(np.exp((-(x - mean)**2)/(2*variance)))
    return f_sub_x

x_coords = []
y_coords = []
for n in range(10000):
    x = 0.1*n
    f_sub_x = normal_density(171, 6.9**2, x)
    x_coords.append(x)
    y_coords.append(f_sub_x)
     
plt.plot(x_coords, y_coords)
plt.show()

def integration(mean, variance, a, b):
    area = 0
    delt_x = (b-a)/10000
    x = a
    for n in range(10000):
        x = a + delt_x*n
        f_X = normal_density(mean, variance, x)
        area += f_X*delt_x
    return area

print(integration(171, 6.9**2, 162, 190))
```
